# Hypothesis

# AutoDrive Components Ltd. – Engine Assembly Time Evaluation

```python
# State Hypothesis
```

```python
# H0
# H1
```

```python
mean = 10.8
theory_mean = 1.8 + 0.02*450
std = 25
sample = 35
alpha = 0.05
```

```python
# t-stats
```

```python
import numpy as np
```

```python
standard_error = std/np.sqrt(sample)
t_statistic = (mean-theory_mean)/standard_error
t_statistic
```

```python
# Critical Value
```

```python
from scipy.stats import norm
critical_value = norm.ppf(1-alpha)
critical_value
```

```python
# Make a Decision
```

if t_statistic>critical_value:
    decision = "H1"
else:
    decision = "H0"

```python
decision
```

```python
# AS per above steps AutoDrive Components Ltd ,there is no change.
```
